```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# Load enriched dataset
df = pd.read_parquet("../../data/processed/incidents_enriched.parquet")
print("Shape:", df.shape)

# 1. Distribution of incident types
plt.figure(figsize=(8, 4))
sns.countplot(data=df, x="incident_type", order=df["incident_type"].value_counts().index)
plt.title("Incident Type Distribution")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 2. Incidents by ride_type
plt.figure(figsize=(10, 5))
top_types = df["ride_type"].value_counts().head(10).index
temp = df[df["ride_type"].isin(top_types)]
sns.countplot(data=temp, y="ride_type", order=top_types)
plt.title("Top 10 Ride Types by Incident Count")
plt.xlabel("Count")
plt.tight_layout()
plt.show()

# 3. Incident type by season
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="season", hue="incident_type")
plt.title("Incident Type by Season")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

# 4. Age group vs. incident type
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x="age_group", hue="incident_type")
plt.title("Incident Type by Age Group")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

# 5. Gender vs. incident type
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x="gender", hue="incident_type")
plt.title("Incident Type by Gender")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

# 6. Heatmap: Day of Week vs. Incident Type
plt.figure(figsize=(8, 5))
crosstab = pd.crosstab(df["day_of_week"], df["incident_type"])
sns.heatmap(crosstab, annot=True, fmt="d", cmap="Blues")
plt.title("Incident Type by Day of Week")
plt.tight_layout()
plt.show()
```
