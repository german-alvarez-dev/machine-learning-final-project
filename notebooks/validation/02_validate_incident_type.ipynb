```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the cleaned dataset
df = pd.read_parquet("../../data/processed/incidents_clean.parquet")

# 1. Check distribution of incident types
incident_counts = df["incident_type"].value_counts()
print("Incident type distribution:")
print(incident_counts)

# 2. Plot bar chart
incident_counts.plot(kind="bar")
plt.title("Distribution of Incident Types")
plt.xlabel("Incident Type")
plt.ylabel("Number of Cases")
plt.tight_layout()
plt.show()

# 3. Inspect most common 'other' descriptions (optional deep dive)
df[df["incident_type"] == "other"]["description"].value_counts().head(10)
```
