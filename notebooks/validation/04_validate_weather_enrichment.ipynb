```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load enriched dataset
df = pd.read_parquet("../../data/processed/incidents_enriched.parquet")

# 1. Check presence of weather columns
weather_cols = ["temperature_max", "temperature_min", "precipitation_sum"]
print("\nðŸ§ª ProporciÃ³n de valores nulos en clima:")
print(df[weather_cols].isna().mean())

# 2. Summary stats
print("\nðŸ“Š EstadÃ­sticas descriptivas:")
print(df[weather_cols].describe())

# 3. Correlaciones climÃ¡ticas
sns.pairplot(df[weather_cols].dropna())
plt.suptitle("DistribuciÃ³n y correlaciÃ³n entre variables climÃ¡ticas", y=1.02)
plt.show()

# 4. Clima por estaciÃ³n del aÃ±o
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x="season", y="temperature_max")
plt.title("DistribuciÃ³n de temperatura mÃ¡xima por estaciÃ³n")
plt.show()

plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x="season", y="precipitation_sum")
plt.title("DistribuciÃ³n de precipitaciÃ³n por estaciÃ³n")
plt.show()

# 5. Clima y tipo de incidente
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x="incident_type", y="temperature_max")
plt.title("Temperatura mÃ¡xima por tipo de incidente")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x="incident_type", y="precipitation_sum")
plt.title("PrecipitaciÃ³n por tipo de incidente")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
